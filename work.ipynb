```python
import os
from dotenv import load_dotenv
load_dotenv()
groq_api=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
llm_model=ChatGroq(model='llama-3.3-70b-versatile', api_key=groq_api)
```

```python
from langchain_community.document_loaders import YoutubeLoader,UnstructuredURLLoader
from langchain.chains import load_summarize_chain
```

```python
url='https://en.wikipedia.org/wiki/Generative_artificial_intelligence'

loader=UnstructuredURLLoader([url.strip()],ssl_verify=False)
data=loader.load()
```

```python
# prompt template
from langchain.prompts import PromptTemplate
prompt_template='''Please summarize the below.
context:{context}'''

prompt=PromptTemplate(template=prompt_template,input_variables=['context'])
```

```python
final_prompt='''Provide the final summary of the entire context with the important points.
Start it with what is actually it means & then provide the summary in number points with example.
context:{context}'''

final_prompt_template=PromptTemplate(input_variables=['context'],template=final_prompt)
```

```python
chain=load_summarize_chain(llm=llm_model,
                           chain_type='map_reduce',
                           map_prompt=prompt,
                           combine_prompt=final_prompt_template,
                           combine_document_variable_name="context",
                           map_reduce_document_variable_name="context",
                           verbose=True)
```

```python
output_summary=chain.run(data)
```
